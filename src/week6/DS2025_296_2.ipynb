```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def plot_height(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    data['theory'] = np.log2(data['n'])

    # Set the style
    sns.set_style("whitegrid")

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(data['n'], data['average'], marker='o')
    plt.plot(data['n'], data['theory'] * data['average'].iloc[-1] / data['theory'].iloc[-1],
             '--', label='O(log n) ref')
    plt.plot(data['n'], data['average'] / data['theory'], ':', label='ratio')
    plt.axhline(y=2, color='r', linestyle='-', alpha=0.7, label='y = 2')
    plt.text(data['n'].iloc[-1] * 0.7, 2.5, 'y = 2', color='r', fontsize=12)
    plt.xlabel('Array Size (n)')
    plt.ylabel('Average Height (ns)')
    plt.title('Average Height vs Array Size')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    
    # Show the plot
    plt.show()
```

```python
plot_height('output_height.csv')
```
