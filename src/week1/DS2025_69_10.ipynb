# <span style='color:gold;'> Plot runtime data</span>

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def plot_runtime(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    data['theory'] = np.log2(data['n'])

    # Set the style
    sns.set_style("whitegrid")

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(data['n'], data['average_time(ns)'], marker='o')
    plt.plot(data['n'], data['theory'] * data['average_time(ns)'].iloc[-1] / data['theory'].iloc[-1],
             '--', label='O(log n) ref')
    plt.xlabel('Array Size (n)')
    plt.ylabel('Average Search Time (ns)')
    plt.title('Average Search Time vs Array Size')
    plt.grid(True, which="both", ls="--", linewidth=0.5)
    
    # Show the plot
    plt.show()
```

```python
# Call the function with the path to the CSV file
plot_runtime('data/runtime.csv')
```
